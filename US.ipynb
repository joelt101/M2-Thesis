```python
import numpy as np
import pandas as pd
```

```python
GSPC = pd.read_excel('/Users/joelthomas/Library/CloudStorage/OneDrive-Personal/Documents/My Documents/UNIVERSITY/Masters/M2 Finance/Master Thesis/Data/US/GSPC.xlsx')
BCOM = pd.read_excel('/Users/joelthomas/Library/CloudStorage/OneDrive-Personal/Documents/My Documents/UNIVERSITY/Masters/M2 Finance/Master Thesis/Data/US/BCOM.xlsx')
TNX = pd.read_excel('/Users/joelthomas/Library/CloudStorage/OneDrive-Personal/Documents/My Documents/UNIVERSITY/Masters/M2 Finance/Master Thesis/Data/US/TNX.xlsx')
USDX = pd.read_excel('/Users/joelthomas/Library/CloudStorage/OneDrive-Personal/Documents/My Documents/UNIVERSITY/Masters/M2 Finance/Master Thesis/Data/US/USDX.xlsx')
```

```python
GSPC.columns
```

```python
dataframes = {
    'GSPC': GSPC,
    'BCOM': BCOM,
    'TNX': TNX,
    'USDX': USDX
}

for name, df in dataframes.items():
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
```

```python
BCOM
```

```python
for name, df in dataframes.items():
    # Replace '-' with NaN in the specified columns
    df.replace('-', np.nan, inplace=True)
    
    # Drop rows with NaN in any of the specified columns
    df.dropna(subset=['Adj Close', 'High', 'Low'], inplace=True)

    # Update the dictionary with the modified DataFrame (optional if inplace=True is used)
    dataframes[name] = df
```

```python
for name, df in dataframes.items():
    # Convert all columns to float
    df = df.astype(float)
    # Update the dictionary with the modified DataFrame
    dataframes[name] = df
```

```python
GSPC
```

```python
GSPC['GSPC_vol'] = (1 / (4 * np.log(2))) * (np.log(GSPC['High']) - np.log(GSPC['Low']))**2
GSPC['GSPC_annvol'] = 100 * np.sqrt(365 * GSPC['GSPC_vol'])
GSPC['GSPC_ret'] = np.log(GSPC['Adj Close'] / GSPC['Adj Close'].shift(1))
GSPC['GSPC_logvol'] = np.log(GSPC['GSPC_vol'])
GSPC['GSPC_logannvol'] = np.log(GSPC['GSPC_annvol'])


BCOM['BCOM_vol'] = (1 / (4 * np.log(2))) * (np.log(BCOM['High']) - np.log(BCOM['Low']))**2
BCOM['BCOM_annvol'] = 100 * np.sqrt(365 * BCOM['BCOM_vol'])
BCOM['BCOM_ret'] = np.log(BCOM['Adj Close'] / BCOM['Adj Close'].shift(1))
BCOM['BCOM_logvol'] = np.log(BCOM['BCOM_vol'])
BCOM['BCOM_logannvol'] = np.log(BCOM['BCOM_annvol'])

TNX['TNX_vol'] = (1 / (4 * np.log(2))) * (np.log(TNX['High']) - np.log(TNX['Low']))**2
TNX['TNX_annvol'] = 100 * np.sqrt(365 * TNX['TNX_vol'])
TNX['TNX_ret'] = TNX['Adj Close'] * 0.01
TNX['TNX_logvol'] = np.log(TNX['TNX_vol'])
TNX['TNX_logannvol'] = np.log(TNX['TNX_annvol'])

USDX['USDX_vol'] = (1 / (4 * np.log(2))) * (np.log(USDX['High']) - np.log(USDX['Low']))**2
USDX['USDX_annvol'] = 100 * np.sqrt(365 * USDX['USDX_vol'])
USDX['USDX_ret'] = np.log(USDX['Adj Close'] / USDX['Adj Close'].shift(1))
USDX['USDX_logvol'] = np.log(USDX['USDX_vol'])
USDX['USDX_logannvol'] = np.log(USDX['USDX_annvol'])
```

```python
annvol = pd.DataFrame(index=dataframes['GSPC'].index)
annvol['GSPC_annvol'] = GSPC['GSPC_annvol']
annvol['BCOM_annvol'] = BCOM['BCOM_annvol']
annvol['TNX_annvol'] = TNX['TNX_annvol']
annvol['USDX_annvol'] = USDX['USDX_annvol']

logvol = pd.DataFrame(index=dataframes['GSPC'].index)
logvol['GSPC_logvol'] = GSPC['GSPC_logvol']
logvol['BCOM_logvol'] = BCOM['BCOM_logvol']
logvol['TNX_logvol'] = TNX['TNX_logvol']
logvol['USDX_logvol'] = USDX['USDX_logvol']

logannvol = pd.DataFrame(index=dataframes['GSPC'].index)
logannvol['GSPC_logannvol'] = GSPC['GSPC_logannvol']
logannvol['BCOM_logannvol'] = BCOM['BCOM_logannvol']
logannvol['TNX_logannvol'] = TNX['TNX_logannvol']
logannvol['USDX_logannvol'] = USDX['USDX_logannvol']


ret = pd.DataFrame(index=dataframes['GSPC'].index)
ret['GSPC_ret'] = GSPC['GSPC_ret']
ret['BCOM_ret'] = BCOM['BCOM_ret']
ret['TNX_ret'] = TNX['TNX_ret']
ret['USDX_ret'] = USDX['USDX_ret']
```

```python
annvol = annvol.replace(0, pd.NA)
logvol = logvol.replace(0, pd.NA)
logannvol = logannvol.replace(0, pd.NA)
```

```python
annvol.to_excel('/Users/joelthomas/Library/CloudStorage/OneDrive-Personal/Documents/My Documents/UNIVERSITY/Masters/M2 Finance/Master Thesis/Data/US/annvol.xlsx')
ret.to_excel('/Users/joelthomas/Library/CloudStorage/OneDrive-Personal/Documents/My Documents/UNIVERSITY/Masters/M2 Finance/Master Thesis/Data/US/ret.xlsx')
logvol.to_excel('/Users/joelthomas/Library/CloudStorage/OneDrive-Personal/Documents/My Documents/UNIVERSITY/Masters/M2 Finance/Master Thesis/Data/US/logvol.xlsx')
logannvol.to_excel('/Users/joelthomas/Library/CloudStorage/OneDrive-Personal/Documents/My Documents/UNIVERSITY/Masters/M2 Finance/Master Thesis/Data/US/US/logannvol.xlsx')
```
